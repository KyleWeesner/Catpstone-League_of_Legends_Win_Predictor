# Model Building

import libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, StackingClassifier
from sklearn.metrics import plot_confusion_matrix, recall_score, accuracy_score, precision_score, f1_score
from sklearn.dummy import DummyClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('../data/full.csv', index_col=0)
df
```

```python
sns.heatmap(abs(df.corr()))
```

```python
df.info()
```

```python
df = df.drop(columns='matchId')
```

```python
X = df.drop(columns='win') #removing the target
y = df.win #isolating the target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3) #performing train test split
```

```python
y.value_counts(normalize=True)
```

```python
subpipe = Pipeline(steps=[('ss', StandardScaler())])
ct = ColumnTransformer(transformers= [
    ('subpipe', subpipe, X.columns)])
```

```python
pipe_dummy= Pipeline(steps=[('ct',ct),
                            ('dummy', DummyClassifier())])

pipe_dummy.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_dummy, X_train, y_train).mean()} , train = {pipe_dummy.score(X_train, y_train)}')
```

```python
pipe_log= Pipeline(steps=[('ct',ct),
                            ('logreg',LogisticRegression(random_state=3))])

pipe_log.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_log, X_train, y_train).mean()} , train = {pipe_log.score(X_train, y_train)}')
```

```python
penalty='l2',
    *,
    dual=False,
    tol=0.0001,
    C=1.0,
    fit_intercept=True,
    intercept_scaling=1,
    class_weight=None,
    random_state=None,
    solver='lbfgs',
    max_iter=100,
    multi_class='auto',
    verbose=0,
    warm_start=False,
    n_jobs=None,
    l1_ratio=None,
```

```python
parameters = {'logreg__penalty': ['l1', 'l2', 'elasticnet', 'none'],
             'logreg__solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}


logreg_gs = GridSearchCV(estimator= pipe_log,
                 param_grid=parameters,
                 cv=5)

logreg_gs.fit(X_train, y_train)
```

```python
logreg_gs.best_params_
```

```python
logreg_gs.best_score_
```

```python
pipe_log= Pipeline(steps=[('ct',ct),
                            ('logreg',LogisticRegression(penalty='l1',
                                                         solver='saga',
                                                         random_state=3))])
parameters = {'logreg__C': [0.1,0.25,.5],
             'logreg__max_iter': [200, 500, 1000]}
             


logreg_gs = GridSearchCV(estimator= pipe_log,
                 param_grid=parameters,
                 cv=5)

logreg_gs.fit(X_train, y_train)
logreg_gs.best_params_ , logreg_gs.best_score_
```

```python
pipe_logreg1 = Pipeline(steps=[('ct',ct),
                            ('logreg', LogisticRegression(C= 0.1,
                                 max_iter= 200,
                                 penalty= 'l1',
                                 solver= 'saga'))])
pipe_logreg1.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_logreg1, X_train, y_train).mean()} , train = {pipe_logreg1.score(X_train, y_train)}')
```

```python
pipe_knn= Pipeline(steps=[('ct',ct),
                            ('knn',KNeighborsClassifier())])

pipe_knn.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_knn, X_train, y_train).mean()} , train = {pipe_knn.score(X_train, y_train)}')
```

```python
n_neighbors=5,
    *,
    weights='uniform',
    algorithm='auto',
    leaf_size=30,
    p=2,
    metric='minkowski',
    metric_params=None,
    n_jobs=None,
    **kwargs,
```

```python
pipe_knn= Pipeline(steps=[('ct',ct),
                            ('knn',KNeighborsClassifier())])

parameters = {'knn__n_neighbors': [3,5,10,25],
             'knn__metric': ['manhattan','euclidean','minkowski']}
             


pipe_knn = GridSearchCV(estimator= pipe_knn,
                 param_grid=parameters,
                 cv=5)

pipe_knn.fit(X_train, y_train)
pipe_knn.best_params_ , pipe_knn.best_score_
```

```python
pipe_knn1= Pipeline(steps=[('ct',ct),
                            ('knn',KNeighborsClassifier())])

parameters = {'knn__n_neighbors': [25,50,100]}
             


pipe_knn1 = GridSearchCV(estimator= pipe_knn1,
                 param_grid=parameters,
                 cv=5)

pipe_knn1.fit(X_train, y_train)
pipe_knn1.best_params_ , pipe_knn1.best_score_
```

```python
pipe_dtc= Pipeline(steps=[('ct',ct),
                            ('dtc',DecisionTreeClassifier(random_state=3))])

pipe_dtc.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_dtc, X_train, y_train).mean()} , train = {pipe_dtc.score(X_train, y_train)}')
```

```python
pipe_rfc= Pipeline(steps=[('ct',ct),
                            ('rfc',RandomForestClassifier(random_state=3))])

pipe_rfc.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_rfc, X_train, y_train).mean()} , train = {pipe_rfc.score(X_train, y_train)}')
```

```python
pipe_rfc= Pipeline(steps=[('ct',ct),
                            ('rfc',ExtraTreesClassifier(random_state=3))])

pipe_rfc.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_rfc, X_train, y_train).mean()} , train = {pipe_rfc.score(X_train, y_train)}')
```

```python
pipe_gbc= Pipeline(steps=[('ct',ct),
                            ('gbc', GradientBoostingClassifier(random_state=3))])

pipe_gbc.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_gbc, X_train, y_train).mean()} , train = {pipe_gbc.score(X_train, y_train)}')
```

```python
stack = StackingClassifier(estimators=[
    ('logreg',LogisticRegression(C= 0.25,
                                 max_iter= 200,
                                 penalty= 'l1',
                                 solver= 'saga')),
    ('knn',KNeighborsClassifier(n_neighbors=50)),
    ('gbc', GradientBoostingClassifier(max_depth= 5, 
                                       min_samples_leaf= 3, 
                                       min_samples_split= 10))])


pipe_stack = Pipeline(steps=[('ct',ct),
                            ('stack', stack)])
pipe_stack.fit(X_train, y_train)
print(f'cv = {cross_val_score(pipe_stack, X_train, y_train).mean()} , train = {pipe_stack.score(X_train, y_train)}')
```
