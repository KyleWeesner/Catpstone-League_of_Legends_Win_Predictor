```python
kills = []
def getKills(game):
    kills = []
    for frame in game['info']['frames'][0:16]:
        for event in frame['events']:
            if event['type'] == 'CHAMPION_KILL':
                kills.append(event) 
    return kills
```

```python
kills_df = pd.DataFrame(getKills(time))
kills_df
```
